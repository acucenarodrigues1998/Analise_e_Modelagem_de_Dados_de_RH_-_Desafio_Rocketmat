### Data Cleaning
A primeira análise a ser feita, depois de conhecida a base de dados, é a verificação de integridade e completude dos dados. Inicialmente, eu criei uma função que retorna um dataframe com os tipos de dado de cada coluna com dados faltantes e a quantidade de dados faltantes. A partir desse dataframe, é possível estudar as variáveis com dados faltantes e definir quais serão as medidas tomadas afim de não perder dados importantes e ter dados integros para as análises que serão realizadas mais a frente. 

```python
# Criação de dataframe com informações de missing data
def verifica_faltantes(data):
    df_faltantes = pd.DataFrame({'tipos': data.dtypes,
                                 'qtd_faltantes': data.isna().sum(),
                                 'percentual_faltante': data.isna().sum() / data.shape[0]},
                                  index=data.columns)
    df_faltantes.sort_values('percentual_faltante', ascending=False, inplace=True)
    return df_faltantes[df_faltantes['percentual_faltante'] > 0]

verifica_faltantes(data)
```

Considerando as colunas com mais de 20% dos dados faltantes, obtemos as seguintes colunas:
* Rising_Star: Indica o nível de promessa ou capacidade de promoção que o empregado tem.
* Critical: O funcionário é crítico para a organização?
* CSR Factor: Ignore
* Women_Leave: Mulheres que deixaram a empresa.
* Men_Leave: Homens que deixaram a empresa

A primeira ação vai ser remover a coluna **CSR Factor** e, juntamente com ela, a coluna **sales**, pois, na descrição das variáveis foi indicado que elas devem ser ignoradas neste estudo. 

Em seguida passamos a verificar a variável **Critical**. Essa variável aparenta ser importantante, entretanto, os dados que nela constam e não são nulos são todos iguais a 1. Uma possível ação seria considerar todas as entradas nulas como 0, mas não é possível afirmar de certeza se todos esses valores são 0. Inputar com estatísticas não seria válido pois, todos os valores seriam 1. Sem falar que essa transformação poderia influenciar negativamente nas aplicações de machine learning. Por esse motivo, essa coluna também será eliminada.

A coluna **Rising_Star** possui quase 100% dos dados faltantes, contando com apenas 21 registros dentro de um universo de quase 15 mil amostras, além do que, esses dados não me dizem muita coisa sobre o valor dessa feature. Essa coluna também será eliminada. 

As colunas **Men_Leave** e **Women_Leave** serão eliminadas para que o gênero não influencie nos resultados, tanto da análise como dos algoritmos de machine learning, uma vez que nos últimos tempos muito se tem discutido sobre a igualdade de gênero. Por este mesmo motivo, a coluna **Gender** também será eliminada (Neste caso, estou considerando que a empresa não possui nenhuma política de incentivo a presença feminina ou qualquer outro tipo de política do tipo).


Passando agora para as variáveis com menos de 20% de completude, temos 5 variáveis como a descrita abaixo:
* EMP_Sat_OnPrem_*: Um indicador de uma pesquisa que foi enviada aos empregados por um terceiro. Esse valor significa que o empregado mantém uma elevada percentagem de trabalho nos locais físicos de trabalho da empresa.

Essas variáveis são importantes pois tratam da porcentagem de trabalho do funcionário no ambiente físico da empresa e pode servir para uma futura análise da produtividade remoto vs empresa. Como se tratam de poucos registros (sendo igual para as 5 variáveis) e após uma pequena análise, foi percebido que são dos mesmos funcionários (que provavelmente não responderam a pesquisa), serão eliminados apenas esses registros nulos. 

Outras variáveis que serão eliminadas são as variáveis **ID** e **Name**, pois, são variáveis que não acrescentam a análise. O ID é gerado automaticamente pelo Pandas e o nome não deve influenciar na análise.

Abaixo são aplicadas as transformações e é verificado se ainda há algum dado faltando.

```python
# Aplicação das transformações propostas acima:
data.drop(['CSR Factor', 'sales', 'Critical', 'Rising_Star', 'Men_Leave', 'Women_Leave', 'Gender', 'ID', 'Name'], axis=1, inplace=True)
data.dropna(subset=['EMP_Sat_OnPrem_1', 'EMP_Sat_OnPrem_2', 'EMP_Sat_OnPrem_3', 'EMP_Sat_OnPrem_4','EMP_Sat_OnPrem_5'], inplace=True)
verifica_faltantes(data)
```

```python
# Shape dos dados após a tranformação
data.shape
```

Uma outra transformação a ser proposta é a transformação das colunas obtidas por pesquisas em uma coluna com a média dos valores de cada pesquisa. Isso vai facilitar o trabalho, uma vez que essas colunas no geral possuem o mesmo significado. Para tal, foi criada uma função que seleciona as colunas com base em um padrão, calcula a média e elimina as colunas que já não são mais úteis.

```python
# Alteração no nome da coluna para adequá-la a um padrão
data.rename(columns = {'Emp_Work_Status2': 'Emp_Work_Status_2'}, inplace=True)
```

```python
def mean_survey_variables(data, regex):
    colunas = data.loc[:, data.columns.str.startswith(regex)].columns
    data[regex+'Mean'] = data.loc[:, colunas].mean(axis=1)
    data.drop(colunas, axis=1, inplace=True)
    
survey_var = ['EMP_Sat_OnPrem_', 'EMP_Sat_Remote_', 'EMP_Engagement_', 'Emp_Work_Status_', 'Emp_Competitive_', 'Emp_Collaborative_']
for var in survey_var:
    mean_survey_variables(data, var)
```

### Correlações entre as variáveis

A partir deste ponto serão analisadas as correlações entre os dados afim de identificar como os dados se relacionam e como uma variável influencia em outra. Essa correlação é calculada utilizando o coeficiente de pearson que assume valores entre -1 e 1. Para a interpretação, quão mais distante de 0, melhor a correlação entre as variáveis, podendo ser negativa ou positiva. 

```python
corr = data.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(15, 12))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True, annot=True)
```

O excesso de informação dificulta um pouco a visualização das correlações dos dados, mas nos dá uma noção de quais variáveis podem ser eliminadas. Essas variáveis são eliminadas abaixo, nos dando uma melhor visualização das principais correlações e do que é relevante pra esse estudo.

```python
data.drop(['Will_Relocate', 'Emp_Collaborative_Mean', 'Emp_Work_Status_Mean', 'Emp_Competitive_Mean', 'EMP_Sat_OnPrem_Mean', 'Work_accident'], axis=1, inplace=True)
corr = data.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(15, 12))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True, annot=True)
```
